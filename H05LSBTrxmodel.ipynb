```python
p1=np.polyfit(H05LSB.lofreq, H05LSB.trx, 1)
p2=np.polyfit(H05LSB.lofreq, H05LSB.trx, 2)
p3=np.polyfit(H05LSB.lofreq, H05LSB.trx, 3)
p4=np.polyfit(H05LSB.lofreq, H05LSB.trx, 4)
```

```python
print(p1)
```

```python
print(p2)
```

```python
print(p3)
```

```python
plt.plot(H05LSB.lofreq, H05LSB.trx,'o')
xp=np.linspace(335, 355, 100)
plt.plot(xp, np.polyval(p1, xp),'r-')
plt.plot(xp, np.polyval(p2, xp),'b--')
plt.plot(xp, np.polyval(p3, xp),'m:')
plt.plot(xp, np.polyval(p4, xp),'k:')
```
